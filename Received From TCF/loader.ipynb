```python
import pandas as pd
import requests
```

```python
data = pd.read_csv('Intermediate institute list.csv', encoding='ISO-8859-1')

data = data.drop(columns=['CODE', 'Tier', 'Institure Type'])
data = data.rename(columns={
    'Institute Name': 'name',
    'Managing Authority': 'managing_authority',
    'Latitude': 'latitude',
    'Longitude': 'longitude',
    'Description': 'description',
    'Gender': 'gender',
    'Faculty': 'faculties',
})
data = data.dropna(subset=['name','latitude', 'longitude'])
```

```python
data['managing_authority'] = data['managing_authority'].str.lower()
data['managing_authority'] = data['managing_authority'].replace(
    to_replace=r'.*government.*',
    value=1,
    regex=True
)
data['managing_authority'] = data['managing_authority'].replace(
    to_replace=r'.*private.*',
    value=2,
    regex=True
)
```

```python
data['gender'] = data['gender'].str.lower()
data['gender'] = data['gender'].replace(
    to_replace=r'.*co.*',
    value=3,
    regex=True
)
data['gender'] = data['gender'].replace(
    to_replace=r'.*female.*',
    value=2,
    regex=True
)
data['gender'] = data['gender'].replace(
    to_replace=r'.*male.*',
    value=1,
    regex=True
)
```

```python
def convert_faculties(faculty):
    if pd.isna(faculty):
        return []
    faculty = faculty.split(',')
    new_faculty = []
    for f in faculty:
        f = f.strip().upper()
        new_f = ""
        for i in f:
            if i.isalpha():
                new_f += i
            else:
                new_f += '-'
        new_faculty.append(new_f)
    return new_faculty
data['faculties'] = data['faculties'].apply(convert_faculties)
data['faculties'] = data['faculties'].apply(lambda x: [f for f in x if f != ''])
```

```python
data['description'] = data['description'].str.strip() if data['description'].notna().any() else ''
```

```python
def create_object(row):
    obj = {
        "name": row['name'],
        "managing_authority": row['managing_authority'],
        "latitude": row['latitude'],
        "longitude": row['longitude'],
        "faculties": [],
        "tcf_rating": 0.0,
        "description": row['description']
    }
    for faculty in row['faculties']:
        obj['faculties'].append({
            "name": faculty,
            "gender": row['gender']
        })

    return obj
```
